# super market billing system 

```python
# Dictionary of available products and their prices
products = {
    "apple": 82,
    "mango": 110,
    "banana": 50,
    "grapes": 90,
    "orange": 60,
    "tomato": 25,
    "potato": 30,
    "onion": 40,
    "broccoli": 120,
    "carrot": 45,
    "milk": 56,
    "cheese": 150,
    "yogurt": 60,
    "butter": 200,
    "bread": 40,
    "eggs": 72,
    "rice": 60,
    "flour": 45,
    "sugar": 44,
    "salt": 20,
    "coffee": 250,
    "tea": 150,
    "soda": 35,
    "juice": 90,
    "water_bottle": 20,
    "chips": 30,
    "biscuits": 25,
    "soap": 45,
    "shampoo": 120,
    "toothpaste": 80,
    "detergent": 70
}

# Main loop to process multiple customers
while True:
    amount = 0
    cart = {}

    # Get customer details
    name = input("Enter customer name: ")
    phone_number = input("Enter phone number: ")

    print("Enter product name and quantity:")

    # Inner loop for processing items in the customer's cart
    while True:
        p_name = input("Enter product name: ").lower()

        if p_name not in products:
            print(f"Sorry, {p_name} is not available.")
            continue

        quantity = float(input("Enter quantity: "))
        amount += products[p_name] * quantity
        cart[p_name] = (products[p_name], quantity)

        repeat_item = input("Do you want to add more items? (yes/no): ").lower()
        if repeat_item in ["no", "n"]:
            break

    # Apply discount based on total amount
    if amount <= 500:
        discount = 0.05
    elif amount <= 1000:
        discount = 0.07
    elif amount <= 5000:
        discount = 0.15
    else:
        discount = 0.2

    act_amount = amount - discount * amount

    # Print receipt
    print("=" * 40)
    print(f"Customer Name: {name}")
    print(f"Phone Number: {phone_number}")
    print("\n========== SUPERMARKET RECEIPT ==========")
    print(f"{'Item':<15}{'Rate':<10}{'Quantity':<10}{'Total'}")
    print("-" * 40)

    for product_name, (price, quantity) in cart.items():
        total_price = price * quantity
        print(f"{product_name.capitalize():<15}₹{price:<10}x {quantity:<8}= ₹{total_price}")

    print("-" * 40)
    print(f"{'Total Bill:':<30}₹{amount:.2f}")
    print(f"{'Discount Applied:':<30}{discount * 100}% off")
    print(f"{'Final Bill:':<30}₹{act_amount:.2f}")
    print("=" * 40)

    another_customer = input("Do you want to enter another customer? (yes/no): ").lower()
    if another_customer in ["no", "n"]:
        print("Thank you for using the Supermarket Billing System. Goodbye!")
        break
```
